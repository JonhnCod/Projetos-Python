```python

import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

df = pd.read_csv("top10s.csv", encoding="ISO-8859-1")
```

```python
top_genre = (df.groupby('top genre')['title'].count().to_frame().reset_index())

fig=px.bar(top_genre,x='top genre',y='title')
fig.show()
```
