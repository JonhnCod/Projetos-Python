```python
dicionario = {}

def menu():
    print("1 - Adicionar item")
    print("2 - Remover item")
    print("3 - Ver lista")
    print("4 - Sair")
    opcao = int(input("Escolha uma Opcao: "))
    return opcao

def executar(escolha):
    if escolha == 1:
        item = input("Digite um item: ").upper()
        qtde = float(input("Digite a quantidade: "))
        dicionario[item] = qtde
    elif escolha == 2:
        item = input("Digite um item: ").upper()
        
        if item in dicionario:
            dicionario.pop(item)
        else:
            print("Item nao encontrado na lista...")
    elif escolha == 3:
        if len(dicionario) == 0:
            print("Lista vazia...")
        else:
            print(dicionario)
    else:
        return 

def main():
    while True:
        digito = menu()
        if digito == 4:
            break
        else:
            executar(digito)
main()
```

```python
previsao_dic = {}

def menu():
    item = input("Digite o nome do produto ( ou 'sair' para sair): ").lower()
    if item == 'sair':
        return False
    else:
        while True:
            try:
                vendas_atual = float(input("Digite as vendas do mes atual: "))
                porcentagem = float(input("Digite a taxa de crescimento (%): "))
                break
                
            except:
                print("Digite apenas Numeros")
        return (item ,vendas_atual ,porcentagem)

def caucular(venda,taxa):
    previsao =  venda + ((taxa/100)*venda)
    return previsao


def adicionar(produto,valor):
    previsao_dic[produto] = valor



def main():
    while True:
        opcao = menu()
        if opcao == False:
            break
        else:
            produto, vendas, taxa = opcao
            valor_crescido = caucular(vendas,taxa)
            adicionar(produto,valor_crescido)
    for i in previsao_dic:
        print(f'{i}: Previsao de vendas do proximo mes = R$ {previsao_dic[i]}')


main()
```

```python
funcionario = {}

def menu():
    vendedor = input("Digite o nome do Vendedor (ou 'sair' para sair): ")
    if vendedor == 'sair':
        return False
    else:
        while True:
            try:
                vendas = int(input("Digite as Vendas: "))
                return (vendedor,vendas)
            except:
                print("valor invalido")

def adicionar(nome,vendas,qtd):
    funcionario[nome] = {'total_vendas':vendas,'quantidade_vendas':qtd}
    print(funcionario)
    print("Adicionado com sucesso...")

def atualizar(nome,venda):
    valores = funcionario[nome]
    vendas_total = valores['total_vendas']
    qtd_total = valores['quantidade_vendas']
    funcionario[nome] = {'total_vendas':vendas_total + venda,'quantidade_vendas':qtd_total + 1}

def descompacatar(nome):
    valores = funcionario[nome]
    vendas_total = valores['total_vendas']
    qtd_total = valores['quantidade_vendas']
    return vendas_total,qtd_total
    

def calcular_media(total_vendas,qtd_vendas):
    media = total_vendas/qtd_vendas
    return media
    
def executar():
    while True:
        opcao = menu()
        if opcao == False:
            break
        else:
            nome_vendedor,vendas = opcao
            if nome_vendedor in funcionario:
                atualizar(nome_vendedor,vendas)

            else:
                adicionar(nome_vendedor,vendas,1)
    

            
executar()
for i in funcionario:
    vendas_total , qtd_vendas = descompacatar(i)
    mediaa = calcular_media(vendas_total,qtd_vendas)
    print(f'{i}: Total de vendas = R$ {vendas_total}, media de vendas = R$ {mediaa}')
```

```python
clientes = {}

segmentos = [(1000,'bronze'), (5000,'Prata'),(float('inf'),'ouro')]

def menu():
    nome = input("Digite o nome do cliente (ou 'sair' para sair): ")
    if nome != 'sair':
        while True:
            try:
                vendas = float(input("Digite o total de compras: "))
            
                return nome,vendas
            except:
                print("valor digitado incorreto...")
    else:
        return False

def segmentar(vendas):
    for i in segmentos:
        limite,segmento = i
        if vendas <= limite:
            return segmento


def adicionar(nome,segmento):
    clientes[nome] = segmento


def executar():
    while True:
        opcao = menu()
        if opcao == False:
            break
        else:
            nome_cliente,vendas_total = opcao
            adicionar(nome_cliente,segmentar(vendas_total))
            print(clientes)

executar()
```

```python
frequencia_letras = {}
frequencia_palavras = {}

def analisador_texto(texto):
    novo_texto = texto.lower()
    palavras = novo_texto.split()
    num_palavras = len(palavras)
    for palavra in palavras:
        if palavra in frequencia_palavras:
            frequencia_palavras[palavra]+=1
        else:
            frequencia_palavras[palavra] = 1
    for ch in novo_texto:
        if ch != " ":
            if ch in frequencia_letras:
                frequencia_letras[ch]+=1
            else:
                frequencia_letras[ch] = 1

    print(f'Contagem de palavras: {num_palavras}')
    print(f'Frequencia de palavras: {frequencia_palavras}')
    print(f'Frequencia de letras: {frequencia_letras}')

analisador_texto("Olá mundo! Este é um teste. Olá novamente.")
```
